```python
number = 100

is_prime = None

if number <= 1: is_prime = False

else:
  for i in range(2 , number):
    if number % i == 0:
      is_prime = True
      break

if is_prime: print('Prime')
else: print('Not Prime')
```

```python
def reverse_string(text:str):
  return text[::-1]

print(reverse_string('Ali'))
print(reverse_string('Ammar'))
```

```python
def count_vowels(text:str):
  vowels = 'aeiou'
  return sum(1 for char in text.lower() if char in vowels)
```

```python
count_vowels('Abu Bakr')
```

```python
count_vowels('Abbas')
```
