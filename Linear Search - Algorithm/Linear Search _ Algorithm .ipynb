```python
def linear_search(numbers: list , target: int) -> int:

  for idx in range(len(numbers)):

    if numbers[idx] == target: return idx

  else: return -1
```

```python
print(linear_search([2 , 6 , 7 , 1 , 12] , 12))
print(linear_search([1 ,2 ,8 , 9 ,10] , 8))
print(linear_search([1 ,2 ,8 , 9 ,10] , 2))
```
